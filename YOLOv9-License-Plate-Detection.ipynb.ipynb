```python
# Import Necessary Libraries:
import torch
import cv2
import os
from google.colab import drive
import utilities as ut
```

```python
# Load YOLOv9 Model:
from ultralytics import YOLO

# Load a pre-trained YOLOv9 model using the custom weights file 'yolov9c.pt'
model = YOLO("yolov9c.pt")
model.info()
```

```python
# Mount Google Drive to access the dataset:
drive.mount('/content/drive')

# Define the path to the YAML configuration file for the YOLO model. This file contains dataset information, including paths to images and labels.
data_path = '/content/drive/MyDrive/YOLO/Test1/Data/data.yaml'


# Train the YOLOv9 Model
results = model.train(
    data=data_path,      # Path to the dataset
    epochs=10,           # Number of training epochs
    imgsz=416,           # Image size (input resolution for the model)
    batch=16,            # Batch size for training
    device="cuda",       # Use GPU (CUDA) for training
    amp=True             # Enable Automatic Mixed Precision for faster training
)
```

```python
# Save the trained YOLOv9 model to Google Drive: The model is saved as a .pt file for future inference or fine-tuning
model.save("/content/drive/MyDrive/YOLO/Test1/trained_lprecog_test1_yolov9c.pt")
```
